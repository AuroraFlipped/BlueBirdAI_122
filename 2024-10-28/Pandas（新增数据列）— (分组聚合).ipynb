# 新增列

```python
df1 = pd.read_csv('./data/beijing_tianqi_2018.csv')
df1
```

```python
# 新建一个最高温通过（类似字符串）replace替换℃同时转换将数据类型转换成'int32'并将转化过的'bWendu'数值赋值给'最高温'
```

```python
 df1['最高温'] = df1['bWendu'].str.replace('℃','').astype('int32')
```

```python
df1
```

```python
# 新建一个最低温通过（类似字符串）replace替换℃同时转换将数据类型转换成'int32'，这里的str只是针对于Series
```

```python
df1['最低温'] = df1['yWendu'].str.replace('℃','').astype('int32')
```

```python
df1
```

```python
# 温差就等于df1中的['最高温'] - df1中的['最低温']
```

```python
df1['温差'] = df1['最高温'] - df1['最低温']
```

```python
df1
```

```python
# 新建数据温度类型
```

```python
df1['温度类型'] = '0'
df1
```

```python
# 新建['bWendu']替换旧的['bWendu']，旧的['bWendu']通过replace方法进行替换替换掉'℃'把值给到新的['bWendu']同时转换数据类型为'int32'
```

```python
# df1.loc['bWendu'] = df1['bWendu'].str.replace('℃','').astype('int32')
# df1
```

```python
df1['bWendu'] = df1['bWendu'].str.replace('℃','').astype('int32')
df1
```

```python
df1['yWendu'] = df1['yWendu'].str.replace('℃','').astype('int32')
df1
```

```python
# 创建温度类型函数
```

```python
def get_wendu_type(x):
    if x['bWendu'] > 33:
        return '高温'
    if x['yWendu'] < -10:
        return '低温'
    return'常温'
```

```python
# 必须指出 axis（轴）的方向
```

```python
df1['温度类型'] = df1.apply(get_wendu_type,axis = 1) 
df1
```

# inner left right outer

```python
female = pd.read_excel('./data/female.xlsx')
female
```

```python
male = pd.read_excel('./data/male.xlsx')
male
```

```python
# inner 内连接  带家人的连接
```

```python
male.merge(female,how = 'inner',on = '配偶')
```

```python
# left right 所有男士上台来，有配偶的带上来，无配偶的举个牌（null），左边的都上来
```

```python
male.merge(female,how = 'left',on = '配偶')
```

```python
male.merge(female,how = 'right',on = '配偶')
```

```python
# outer 全连接 所有人都上台来，无配偶举个牌null
```

```python
male.merge(female,how = 'outer',on = '配偶')
```

# 万能公式

```python
pd.merge(male,female,on = '配偶',how = 'inner')
```

```python
pd.merge(male,female,on = '配偶',how = 'right')
```

```python
pd.merge(male,female,on = '配偶',how = 'outer')
```
