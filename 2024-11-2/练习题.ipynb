```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('./data/saleinfo_month.xls')
df
```

```python
data = df[['manufacture','quantity']].groupby('manufacture').agg({'quantity':'sum'}).sort_values(by = 'quantity',ascending = False).head(10)
data
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig = plt.figure(figsize=(20,8))
x = data.index
y = data['quantity']
plt.xlabel('厂商',fontsize=20,color='red')
plt.ylabel('销量',fontsize=20,color='red')
plt.title('各厂商销量对比之TOP10',fontsize=20,color='green')
plt.bar(x,y)
```

```python
data2 = df.groupby(['series','cartype']).agg({'quantity':'sum'}).unstack()
data2
```

```python
data2.plot(kind = 'bar',stacked = True)
plt.title('相同车系不同车型销量对比分析',fontsize = 15)
plt.xlabel('车系',fontsize = 10)
plt.ylabel('销量',fontsize = 10)
plt.legend(loc = 1)
```
