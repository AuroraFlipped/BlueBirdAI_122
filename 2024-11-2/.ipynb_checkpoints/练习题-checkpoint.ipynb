```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('./data/saleinfo_month.xls')
df
```

```python
df[['厂商','销量']] =df[['manufacture','quantity']]
df
```

```python
df1 = df[['厂商','销量']].groupby('厂商').agg({'销量':'sum'})
df1
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig = plt.figure(figsize = (20,8))
x = df1.index
y = df1['销量']
plt.title('各厂商销量对比')
plt.bar(x,y)
```
