```python
bbb = data1[['籍贯','姓名']].groupby('籍贯').agg({'姓名':'nunique'})
bbb
```

```python
data5 = pd.merge(aaa,bbb,how = 'inner',on = '籍贯')
data5
```

```python
data5.sort_values(by = '姓名',ascending = False)
```

```python
list(data1[['籍贯','X','Y']].groupby('籍贯'))
```

```python
# 由上面的情况得知有重复的可以使用.first()
ccc = data1[['籍贯','X','Y']].groupby('籍贯').first()
ccc
```

```python
ccc['人数'] = data1[['籍贯','姓名']].groupby('籍贯').agg({'姓名':'nunique'})
ccc
```

```python
ccc.sort_values(by = '人数',ascending = False)
```

# 构造经纬度字典

```python
#方法
#1.把数据何存到json文件当中
#定义一个字典保存籍贯和对应的经纬度
address_dict = {}
# 首先获取所有的籍贯
for city in ccc.index.tolist():
    address_dict[city]=ccc.loc[city,['X','Y']].tolist()
address_dict
```

```python
#保存到json文件中
import json
with open('./addr.json','w',encoding='utf-8') as f:
    json.dump(address_dict,f,ensure_ascii=False) # ensure_ascii=False防止中文乱码
```

```python
#复制官网原形，然后更改相应设置
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.faker import Faker

c = (
    Geo(init_opts=opts.InitOpts(width='1200px',height='800px'))
    .add_schema(maptype="china",is_roam=False)
    .add_coordinate_json('./addr.json')
    .add("官员人数", [list(z) for z in zip(ccc.index.tolist(), ccc['人数'].values.tolist())])
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(), title_opts=opts.TitleOpts(title="Geo-基本示例")
    )
#    .render("geo_base.html")

)
c.render_notebook()
```
