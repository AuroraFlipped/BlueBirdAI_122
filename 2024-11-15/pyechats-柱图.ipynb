# 需求：不同年月销售额趋势分析
1. 维度：年月
2. 度量：销售额

4. 不同字样：认为必须是分组

```python
df = pd.read_excel('./超市.xlsx')
df
```

```python
# 查询每列的title
```

```python
df.columns
```

```python
df['订单日期']
```

```python
df['年月'] = df['订单日期'].apply(lambda x:x.strftime('%Y-%m'))
```

```python
data = df[['年月','销售额']].groupby('年月').agg({'销售额':'sum'})
```

```python
data
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker

c = (
    Line()
    .add_xaxis(data.index.tolist())
    .add_yaxis("总销售额", data['销售额'].tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title="不同年月销售额趋势分析",pos_left = 'center',title_textstyle_opts = {'font_size':40,'color':'red'}),xaxis_opts=opts.AxisOpts(name_rotate=90
# font_size没有生效，color生效了从xasix_opts设置x轴文本的倾斜方向name可以重新设置
,axislabel_opts={"rotate":45},name ='月份'))
#     .render("line_base.html")
)
c.render_notebook()
```

# 换一种方式修饰

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker

c = (
    Line()
    .add_xaxis(data.index.tolist())
    .add_yaxis("总销售额", data['销售额'].tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title="不同年月销售额趋势分析",pos_left = 'center',title_textstyle_opts=opts.TextStyleOpts(color="#f00", font_size=40)),xaxis_opts=opts.AxisOpts(name_rotate=90
,axislabel_opts={"rotate":45},name ='月份'))
#     .render("line_base.html")
)
c.render_notebook()
```

# 让600000变成60万，让图形变的圆滑

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker

c = (
    Line()
    .add_xaxis(data.index.tolist())
    # lambda 用来让600000变成60（，2）表示保留两位小数，is_smooth = True 是让图变的更圆滑
    .add_yaxis("总销售额", data['销售额'].apply(lambda x:round(x*0.0001,2)).tolist(),is_smooth = True)
    .set_global_opts(title_opts=opts.TitleOpts(title="不同年月销售额趋势分析",pos_left = 'center',title_textstyle_opts=opts.TextStyleOpts(color="#f00", font_size=40)),xaxis_opts=opts.AxisOpts(name_rotate=90
,axislabel_opts={"rotate":45},name ='月份'))
#     .render("line_base.html")
)
c.render_notebook()
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker

c = (
    Line()
    .add_xaxis(data.index.tolist())
    .add_yaxis("总销售额", data['销售额'].apply(lambda x:round(x*0.0001,2)).tolist(),is_smooth = True)
    # 这个set_series是官方做的在数值后面加单位‘万’，position可以更改值可以改成left
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{@[1]}万',position='right'))
    
    .set_global_opts(title_opts=opts.TitleOpts(title="不同年月销售额趋势分析",pos_left = 'center',title_textstyle_opts=opts.TextStyleOpts(color="#f00", font_size=40)),xaxis_opts=opts.AxisOpts(name_rotate=90
,axislabel_opts={"rotate":45},name ='月份'))
    
#     .render("line_base.html")
)

c.render_notebook()
```

```python
# 不同地区不同类别销售额对比分析
data4 = df[['地区','类别','销售额']].groupby(['地区','类别']).agg({'销售额':'sum'}).reset_index()
data4
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker


c = (
    Bar()
    .add_xaxis(data4['地区'].unique().tolist())
    .add_yaxis("商家A", data4.query("类别 == '办公用品'")['销售额'].tolist(),stack="stack1")
    .add_yaxis("商家B", data4.query("类别 == '家具'")['销售额'].tolist(),stack="stack1")
    .add_yaxis("商家C", data4.query("类别 == '技术'")['销售额'].tolist(),stack="stack1")
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(title_opts=opts.TitleOpts(title="Bar-堆叠数据（部分）"))
#     .render("bar_stack1.html")
)
c.render_notebook()
```
